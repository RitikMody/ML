```python
# Compute Cosine Similarity
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
count_vectorizer = CountVectorizer()
```

```python
def vector(x):
    sparse_matrix=count_vectorizer.fit_transform(x)
    doc_term_matrix = sparse_matrix.todense()
    df = pd.DataFrame(doc_term_matrix)
    result=cosine_similarity(df, df)
    return result[0][1]
```

```python
data['test']=data['description_x']+";"+data['description_y']
```

```python
 data['test']=data['test'].apply(lambda x: x.split(";"))
```

```python
data['similarity %']=100*data['test'].apply(lambda x: vector(x))
```
